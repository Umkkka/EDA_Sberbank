# Описание

ПАО "Сбербанк" - государственная российская банковская и финансовая компания со штаб-квартирой в Москве. Сбербанк осуществляет свою деятельность в ряде европейских и постсоветских стран. По состоянию на 2014 год он был крупнейшим банком в России и Восточной Европе, третьим по величине в Европе, занимал 33-е место в мире и первое место в Центральной и Восточной Европе в рейтинге Banker Top 1000 World Banks.

Набор содержит в себе информацию из 84 регионов России за период с 01.2013 по 01.2019 год:

* **Расходы**
  * *'Average spending in a fast food restaurant', 'Average spending in a restaurant'* - Средние расходы в фаст фуде и в ресторане
  * *'Average Fast Food format Check', 'Average check in Restaurant format'* - Средний чек в фаст фуде и в ресторане

* **Банковские операции**
  * Потребительский кредит
    * *'The number of applications for consumer loans'* - Количество заявок на потребительские кредиты

    * *'The number of applications for mortgages'* - Количество заявок на ипотеку

# Предобработка данных

- Импортируем данные. Для того чтобы их было легко анализировать разобьём параметры по отдельным столбцам

```python
data = pd.read_csv('/kaggle/input/sberbank-open-data-eng/df.csv', index_col = 'region', parse_dates = True)
data.head()
```

```python
data = data.pivot_table(index = ["region", "date"], columns = "name", values = "value", aggfunc = "mean", fill_value = 0).reset_index()
data.head()
```

- Выкидываем общие данные с регионом "Россия", чтобы общие данные по стране не создавали выбросы. Будем анализировать данные по регионам

```python
data.drop(data[data['region'] == 'Russia'].index, inplace = True)

data.shape
```

- Добавим столбцы месяц и год для того, чтобы было легко находить зависимость от определённого периода

```python
data['year'] = pd.DatetimeIndex(data['date']).year 
data['month'] = pd.DatetimeIndex(data['date']).month 
data.drop(['date'], axis = 1, inplace = True)
data.head()
```

- Проверим данные на пропуски - пропуски не найдены

```python
data.isna().sum()[data.isna().sum() > 0]
```

- Проверим данные на нулевые значения. 2013 и 2019 имеют много столбцов с нулевыми значениями. Также исключим данные этих годов во избежании выбросов

```python
data[data['year'] == 2013].describe()
```

```python
data[data['year'] == 2019].describe()
```

```python
data.drop(data[data['year'] == 2013].index, inplace = True)
data.drop(data[data['year'] == 2019].index, inplace = True)

data.shape
```

- Необходимо разбить исходную выборку на три категории: расходы, доходы и банковские операции. Сделаем это для удобства нахождения зависимостей у категорий разных подвыборок. Также переименуем столбцы 

```python
expenses = data[['region', 'month', 'year', 'Average spending in a fast food restaurant', 'Average spending in a restaurant', 
                 'Average Fast Food format Check', 'Average check in Restaurant format', 'Average spending on cards']]
expenses = expenses.rename(columns = {'Average spending in a fast food restaurant': 'aver_fastfood', 'Average spending in a restaurant': 'aver_rest',
                                     'Average Fast Food format Check': 'aver_check_fastfood', 'Average check in Restaurant format': 'aver_check_rest',
                                     'Average spending on cards': 'aver_cards'})
expenses.head()
```

```python
income_savings = data[['region', 'month', 'year', 'Average salary', 'Average pension', 'On average rubles on current account per person', 
                       'On average, deposits in rubles per person']]
income_savings = income_savings.rename(columns = {'Average salary': 'aver_salary', 'Average pension': 'aver_pension',
                                                  'On average rubles on current account per person': 'aver_curacc_perpers',
                                                  'On average, deposits in rubles per person': 'aver_deposits_perpers'})
income_savings.head()
```

```python
bank_operations = data[['region', 'month', 'year', 'The number of applications for consumer loans', 'Average consumer loan application', 
                        'The number of applications for mortgages', 'The average amount of a mortgage application', 
                        'Number of new deposits', 'Average amount of new deposit']]
bank_operations = bank_operations.rename(columns = {'The number of applications for consumer loans': 'num_loan_applic', 
                                          'Average consumer loan application': 'aver_loan_applic', 
                                          'The number of applications for mortgages': 'num_mortgag_applic', 
                                          'The average amount of a mortgage application': 'aver_mortgag_applic',
                                          'Number of new deposits': 'num_new_depos',
                                          'Average amount of new deposit': 'aver_new_depos'})
bank_operations.head()
```

# Визуализация данных

* **Доходы и сбережения**

В первую очередь проанализируем и визуализируем подвыборку с доходами и сбережениями (income_savings). Далее проанализируем, какая есть зависимость у двух других подвыборок от описанной выше.

Первые две диаграммы говорят о том, что среднее количество новых заявок на вышеописанные услуги по всей стране имеет тенденцию к росту. Замечается "просадка" данного показателя в 2015 году. Последующие две диаграммы, отображающие средний размер новых заявок на уже известные услуги по всей стране, также говорят о положительной динамики роста данных показателей с течением времени. Также заметна "просадка" в 2015 году в среднем размере новых заявок на кредит. 

**По построенным диаграммам можно с уверенностью сделать вывод, что население России всё менее скептично настроено к кредитам и ипотекам: растёт количество новых заявок по данным услугам, а также размер суммы по данным заявкам.**

Интересно узнать зависимость некоторых показателей от показателя "Средний расход по карте" и от друг друга. Отберём наиболее интересные признаки и построим между ними корреляционную матрицу. 

Из построенной матрицы можно сделать следующие выводы:
* Показатель "Средние расходы по карте" (aver_cards) коррелирует с показателями "В среднем рублей на расчетном счете на человека" (aver_curacc_perpers) - 0.86 и с показателем "Средний размер потребительского кредита" (aver_loan_applic) - 0.82. Первая зависимость логична - человек тратит столько, сколько у него есть. В свою очередь вторая зависимость говорит о том, что люди не имеют управлять своими тратами - тратят больше, чем имеют, в связи с чем вынуждены брать потребительские кредиты. 
* Показатель "В среднем рублей на расчетном счете на человека" (aver_curacc_perpers) взаимно зависит от показателя "Средний размер потребительского кредита" (aver_loan_applic) - 0.86. Вполне возможно, что данные признаки коррелируют ввиду того, что люди имеют кредитные карты для повседневных покупок до момента пока не будет выплачена заработная плата. Но тогда этот признак также бы зависил и от признака "Средняя заработная плата" (aver_salary), зависимости не наблюдается - 0.59. Скорее всего здесь можно сделать самый логичный вывод - человек берет кредит на определенную сумму, эта сумма находится на его расчетном счете.
* Показатель "В среднем размер вклада на человека" (aver_deposits_perpers) взаимно зависит от "Среднего размера потребительского кредита" (aver_loan_applic) - 0.75. Возможно, что данные признаки коррелируют ввиду того, что чем больше у человека вклад (накоплено средств), тем больше у него возможности и уверенности погасить большую сумму кредита.
* "Количество заявок на потребительские кредиты" (num_loan_applic) взаимно зависят с "Количеством заявок на ипотеку" (num_mortgag_applic) - 0.88. Это еще раз подтверждает одинаковое отношение населения к кредитам и ипотекам.
* "Средний размер потребительского кредита" (aver_loan_applic) коррелирует с "Средним размером ипотеки" (aver_mortgag_applic) - 0.84. Чем больше размер кредита, тем больше размер ипотеки (и наоборот).

```python
new_df = data[['region', 'year', 'Average spending on cards', 'Average salary', 'Average pension', 'On average rubles on current account per person', 
               'On average, deposits in rubles per person', 'The number of applications for consumer loans', 'Average consumer loan application', 
               'The number of applications for mortgages', 'The average amount of a mortgage application']]
new_df = new_df.rename(columns = {'Average spending on cards': 'aver_cards', 'Average salary': 'aver_salary', 'Average pension': 'aver_pension',
                                  'On average rubles on current account per person': 'aver_curacc_perpers', 'On average, deposits in rubles per person': 'aver_deposits_perpers',
                                  'The number of applications for consumer loans': 'num_loan_applic', 'Average consumer loan application': 'aver_loan_applic', 
                                  'The number of applications for mortgages': 'num_mortgag_applic', 'The average amount of a mortgage application': 'aver_mortgag_applic', })
new_df.head(10)
```

```python
plt.figure(figsize = (8, 8));
plt.title('Зависимость показателей из разных подвыборок друг от друга', fontsize = 15)
sns.heatmap(new_df.corr(), annot = True, cmap = "Greens_r");
```

Посчитаем точный коэффициент корреляции Пирсона (PCC) между вышеописанными признаками. Тем самым докажем приведенные выводы и тезисы.

![](http://miro.medium.com/max/436/1*G9-Dwjg-UVB9uHCfI7M4hA.png)

```python
def corr_coeff(df, x, y):
    x_bar = df[x].mean()
    y_bar = df[y].mean()
    
    x_sd = df[x].std()
    y_sd = df[y].std()
    
    total_prod = (((df[x] - x_bar) / x_sd) * ((df[y] - y_bar) / y_sd)).sum()
    corr = total_prod / (df.shape[0] - 1)
    
    return corr
```

```python
print('PCС между средними расходами по карте и средним размром кредита - ', corr_coeff(new_df, 'aver_cards', 'aver_loan_applic'))
print('PCС между средней суммой на расчетном счете на человека и средним размером кредита - ', corr_coeff(new_df, 'aver_curacc_perpers', 'aver_loan_applic'))
print('PCС между средним размером вклада на человека и средним размром кредита - ', corr_coeff(new_df, 'aver_deposits_perpers', 'aver_loan_applic'))
```
